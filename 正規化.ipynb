集合　$ A , B $ があって

$ x \in A $ ,  $ y \in B $ として 



関数 $f$ は　こんな風に書くよね

$ y = f(x) $

$ A $ の要素に対して $ B $ の要素が 1つ決まる 

$n$個の集合 $  \; A_1  \; A_2  \;  A_3  \dots \; A_n $　の直積集合 

$  \; A_1  \times  A_2  \times  A_3  \times \dots \; \times A_n $の 要素

$ ( a_1 , a_2 , a_3 \dots a_n ) $ 

に対しても $ ( a_1 , a_2 , a_3 \dots,  a_n ) $  に対して　$ B $ の要素が 1つ決まれば

$ y = f( a_1 , a_2 , a_3 \dots ,a_n) $

これは $ n $ 引数の関数だよね

|$A$ | $B$|
|--|--|
|$a_1$ | $b_1$|
|$a_2$ | $b_2$|
|$a_3$ | $b_3$|
|$a_4$ | $b_4$|
|$a_5$ | $b_5$|

表全体に対して ある列(の集合) を指定すれば 表のデータの  1行が定まる時に　その列(の集合)のことをキーと呼ぶ

列の集合を指定して 表のデータの  1行が定まる時に 集合の要素を減らしても 同様に　1行が定まることがある
それ以上減らすと、1行が定まらくなるような列の集合のことを 候補キーと呼ぶ

逆に 候補キーの列の集合に他の列を加えても、表のデータの 1行は定まるので、このような列の集合は　スーパーキーと呼ぶ

候補キーの選び方は1つとは限らないので、候補キーの選びからから1つ選んで 主キーとする。
もちろん候補キーが1つしかないなら、それが主キーとなる

#  キー

空の表を作って、名前をつけて、列を組で表し、主キーにアンダーラインを引く
これを リレーショナルスキーマと呼ぶ

### 候補キー
本校の学生名簿を考える。

名簿は　学生番号, 学科、学年、出席番号、氏名、住所が 載っているとしよう

キーとなるのは、その項目を指定すれば、学生が高々1人決まる ということ（複数にならないということ）

ホゲホゲ科の ホゲラ年 出席番号 ホゲ　の学生　と呼び出しをかければ 一人を呼び出しているということになる。
つまり　{学科,学年,出席番号} は キーであり　

このうち1つでも欠ければ 一人を呼び出しているということにならない

> ホゲホゲ科の ホゲラ年の学生来てください。といえば複数の学生がやってくる

ので、{学科,学年,出席番号} は候補キーである。

####  候補キーは 極小
 候補キーは それ以上 項目（列、属性 という言い方もある）を減らすと キーではなくなる。この性質のことを候補キーは 極小であるという。　


### スーパーキー

候補キーの項目を含んでいれば、えておけば、それの別の項目を加えても キーとなる

例えば ホゲホゲ科の ホゲラ年 出席番号 ホゲ の 何某さん　と 氏名を加えても呼びだすことができる.

繰り返しになるが、
{学科,学年,出席番号,氏名} 　はキーであるが　氏名がなくても、呼び出せるのでこれは、候補キーでない。（極小でない）
前述のように {学科,学年,出席番号}は候補キーである、候補キーを含む項目はスーバーキーと呼ばれる。
{学科,学年,出席番号,氏名} 　は スーバーキーである



## 主キー

候補キーの選び方は1通りに 限らないかもしれない、1通りにしか選べない場合は、その候補キー、複数あった場合は,そのなかの1つの選び方を
選んで主キーとする。主キーを指定して、リレーショナルスキーマを記述する。



候補キー  {学科,学年,出席番号} を主キーとして　学生名簿のリレーショナルスキーマを記述すると

学生名簿 ( 学生番号, <u>学科</u>、<u>学年</u>、<u>出席番号</u>、氏名、住所)

となる

一方　{学生番号} を指定しても　　学生が高々1人決まる

>学生番号 ホゲ番の学生 は　一人しかいない

ので {学生番号}も候補キー である　

候補キー  {学生番号} を主キーとして　学生名簿のリレーショナルスキーマを記述すると


学生名簿 ( <u>学生番号</u> 学科、学年、出席番号、氏名、住所)

となる

候補キーが複数選べる時、どれを 主キーにするかの、決まりはない。そのデータベースをどのように利用するかなどによって、
リレーショナルスキーマを定義する人が、決める。
